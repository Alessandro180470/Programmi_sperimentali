```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from random import random


```

```python
np.random.seed(42)
x = np.random.standard_normal((200,1))
```

```python
y = 2 * x + 3
```

```python
plt.plot(x,y)
plt.grid(True)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))
])#modello sequenziale il primo layer deve avere sempre questo parametro input_shape =1 significa che vuole sapere di quante colonne e formato l'input
model.summary()#come possiamo vedere ci sono due parametri W e B dove la W viene moltiplicata mentre la B addizionata
#la W dovrà avvicinarsi a 2 e la B dovrà avvicinarsi a 3
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
model.compile(loss=tf.keras.losses.MeanSquaredError(),
              optimizer=tf.keras.optimizers.Adam())
```

```python
history = model.fit(x=x, y=y, epochs=1000)
```

```python
y_predict = model.predict(x)
```

```python
np.concatenate((y_predict, y), axis=1)
```

```python
mse = tf.keras.losses.MeanSquaredError()
loss_value = mse(y[:,0], y_predict[:,0]).numpy()
print(loss_value)
```

```python
loss_history = history.history['loss']
plt.plot(range(1000),loss_history)
plt.grid(True)
```
